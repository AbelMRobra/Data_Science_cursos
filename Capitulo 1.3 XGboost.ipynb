```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv("base_datos_2008.csv")
```

```python
df = data.dropna(subset = ["ArrDelay"])
df = df.sample(frac = 1).head(10000)

X = df[["AirTime", "Distance", "TaxiIn", "TaxiOut", "DepDelay"]]
Y = df["ArrDelay"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = .2, random_state = 1)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = XGBRegressor(n_jobs=-1, learning_rate=5, max_depth=2, 
                    colsample_bytree=1, verbosity=2, subsample=1, n_estimators=500)
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions)
```
